 # The importance of investing
 A small experiment showing the importance of investing you money.

 ## Settings

 Here are the settings of the simulation.
 We will then simulate three scenarios:
 * just saving money every month
 * Saving money every month with some annual return.
 * Saving money every month with some annual return but taking money for a vacation every couple of years.



```python
# We save 10.000kr every month.
money_per_month = 10000
# We save for 30 years.
nr_years = 30
# We assume that we have an return of 5% every year.
increase_per_year = 1.05
# We start with 0kr
starting_sum = 0
# We take a vacation every 5 years.
vaction_years = 5
# The vacation costs 200000kr
vacation_cost = 200000
```

 ## Calculations

```python
increase_per_month = increase_per_year**(1/12)
sum_with_interest = [starting_sum]
sum_with_interest_and_vacation = [starting_sum]
sum_without_interset = [starting_sum]

for m in range(nr_years * 12):
    sum_with_interest.append(sum_with_interest[-1] * increase_per_month + money_per_month)

    sum_with_interest_and_vacation.append(sum_with_interest_and_vacation[-1] * increase_per_month + money_per_month)

    if m % (vaction_years*12) == (vaction_years*12-1):
        sum_with_interest_and_vacation[-1] -= vacation_cost

    sum_without_interset.append(sum_without_interset[-1] + money_per_month)
```

 ## Plotting the money over time

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
plt.figure(figsize=(14, 8))

x = [i/12 for i in range(len(sum_with_interest))]
plt.plot(x, sum_with_interest, label='Sum with interest')
plt.plot(x, sum_with_interest_and_vacation, label='Sum with interest and vacations')
plt.plot(x, sum_without_interset, label='Sum without interest')
plt.legend(fontsize=18)
```


 ## Printing the result


```python
print(f'Sum with interest: {int(sum_with_interest[-1]):,}')
print(f'Sum with interest and vacations: {int(sum_with_interest_and_vacation[-1]):,}')
print(f'Sum without interest: {int(sum_without_interset[-1]):,}')
```
